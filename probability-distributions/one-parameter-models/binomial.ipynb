## Binomial Distribution

Binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent and identically distributed (i.i.d.) Bernoulli trials. Each trial can result in either a success with probability $p$ or a failure with probability $1-p$. The probability of getting $k$ successes in $n$ independent trials is given by the following binomial probability mass function:

### Characteristics of a Binomial Distribution


1. Two Possible Outcomes: The binomial distribution consists of two possible outcomes (success or failure) for each trial.

2. Fixed Number of Trials: The binomial distribution involves a fixed number of independent trials.

3. Constant Probability of Success: The probability of success remains constant across all trials.

4. Independent Trials: Each trial is independent, meaning that the outcome of one trial does not affect the outcome of another trial.

### Probability Mass Function

The probability mass function (PMF) of a binomial distribution gives the probability of getting k successes in n trials, given the probability of success p for each trial. The PMF is given by the following formula:

\begin{equation}
P(X=k) = {n \choose k} p^k (1-p)^{n-k}
\end{equation}


where:

- n is the number of trials
- k is the number of successes
- p is the probability of success for each trial
- ${n \choose k}$ is the binomial coefficient, which determines the number of ways to choose k successes from n trials


```python
import numpy as np
import matplotlib.pyplot as plt
import math

def binomial_pmf(n, p, k):
    """
    Calculates the probability mass function of the binomial distribution.
    
    n: int - number of trials
    p: float - probability of success
    k: int or list of ints - number of successes

    return: float or np.ndarray - probability of k successes in n trials
    """
    coef = math.comb(n, k)
    pmf = coef * (p ** k) * ((1-p) ** (n-k))
    return pmf
```

```python
# Plot the PMF
p = 0.3
n = 10
k_range = np.arange(0, n+1)

pmf = [binomial_pmf(n, p, k) for k in k_range]
plt.stem(k_range, pmf, use_line_collection=True)
plt.xlabel('K')
plt.ylabel('P(X = K)')
plt.title(f'Binomial PMF with n={n} and p={p}')
plt.show()
```

### Cumulative Distribution Function

The cumulative distribution function (CDF) of the binomial distribution gives the probability that the random variable X takes a value less than or equal to a certain value (k), and is given by:

$F(k) = \sum_{i=0}^{k} {n \choose i}p^i(1-p)^{n-i}$

Where:
- $F(k)$ is the CDF of X
- $n$, $p$ and $k$ have the same meaning as in the PMF.

```python
def binomial_cdf(n, p, k):
    """
    Calculates the cumulative distribution function of the binomial distribution.
    
    n: int - number of trials
    p: float - probability of success
    k: int or list of ints - maximum number of successes to include in the CDF

    return: float or np.ndarray - probability of X <= k for a binomial distribution with parameters n and p
    """
    cdf = np.zeros_like(k, dtype=np.float64)
    for i in range(len(k)):
        for j in range(k[i]+1):
            cdf[i] += math.comb(n, j) * (p ** j) * ((1-p) ** (n-j))
    return cdf
```

```python
# Plot the CDF
cdf = binomial_cdf(n, p, k_range)
plt.step(k_range, cdf, where='post')
plt.xlabel('K')
plt.ylabel('P(X <= K)')
plt.title(f'Binomial CDF with n={n} and p={p}')
plt.show()
```

### Exchangeability

Exchangeability is a concept in probability theory that refers to a situation in which a set of random variables has the same probability distribution regardless of the order of the variables. In other words, if we can permute the order of the variables (i.e., exchange them) without changing the probability distribution, then we say that the variables are exchangeable.

The binomial distribution is an example of an exchangeable distribution. To see why, consider a sequence of independent Bernoulli trials with success probability $p$, and let $X$ be the number of successes in $n$ trials. If we permute the order of the trials, the resulting sequence of successes and failures may look different, but the number of successes $X$ remains the same. In other words, the probability distribution of $X$ does not depend on the order in which the trials were conducted, so the Bernoulli trials are exchangeable.

More generally, the binomial distribution can be thought of as a distribution that arises from a sequence of exchangeable events. Any sequence of events that are exchangeable in this way can be modeled using a binomial distribution, which makes the binomial distribution a very useful tool in modeling a wide variety of real-world problems related to counting events.

It is worth noting that while the binomial distribution is exchangeable, not all discrete distributions are. For example, the Poisson distribution is not exchangeable, since the number of events that occur in a given time interval depends on the length of that interval.


### Relationship with Normal Distribution

The relationship between the binomial distribution and normal distribution can be described by the Central Limit Theorem. According to this theorem, when the sample size is large (typically greater than 30) and the probability of success on each trial is not too close to 0 or 1, the binomial distribution can be approximated by the normal distribution with a certain mean and variance. 

More specifically, if $X$ is a binomially distributed random variable with parameters $n$ and $p$, then the mean of $X$ is given by $\mu = np$ and the variance is given by $\sigma^2 = np(1-p)$. Then, the standardized variable $Z = \frac{X-\mu}{\sigma}$ can be approximately modeled by the standard normal distribution, which has a mean of 0 and a variance of 1. Mathematically:

$$ Z = \frac{X-np}{\sqrt{np(1-p)}} \ \xrightarrow{d} \ \mathcal{N}(0,1) $$

where $\xrightarrow{d}$ denotes convergence in distribution, and $\mathcal{N}(0,1)$ denotes the standard normal distribution.

This approximation can be useful in situations where there are too many possible outcomes to calculate exactly, or where the sample size is too large to analyze using the binomial distribution. It allows us to make probability calculations using the familiar normal distribution instead.


### Binomial and Beta Distribution in Bayesian Inference

The relationship between the binomial distribution and beta distribution is important in Bayesian inference, where the binomial distribution is commonly used to model the likelihood of a series of binary outcomes (such as success or failure), and the beta distribution is used as a prior distribution for the unknown probability of success.

Specifically, if we observe a sequence of binary outcomes $y_1, y_2, ..., y_n$ with $n$ total trials and probability of success $p$, the likelihood function is given by the binomial distribution:

$$P(y_1, y_2, ..., y_n | p) = p^k(1-p)^{n-k}$$

where $k=\sum_{i=1}^ny_i$ is the number of successes. In Bayesian inference, we also specify a prior distribution for the unknown probability of success $p$. A commonly used distribution is the beta distribution, which is a continuous probability distribution on the interval [0,1]. 

The beta distribution has two parameters, $\alpha$ and $\beta$, which control the mean and variance of the distribution. The probability density function of the beta distribution is given by:

$$ f(p| \alpha, \beta) = \frac{p^{\alpha-1}(1-p)^{\beta-1}}{B(\alpha, \beta)} $$

where $B(\alpha, \beta)$ is the beta function, which serves as a normalizing constant to ensure that the integral over the interval [0,1] equals 1. The beta distribution has three common shapes: J-shaped, U-shaped, and bell-shaped.

Using Bayes' theorem, we can update our prior belief about the probability of success given the observed data. The posterior distribution is proportional to the product of the prior and the likelihood function:

$$ P(p | y_1, y_2, ..., y_n) \propto P(y_1, y_2, ..., y_n | p) P(p) $$

Using the beta distribution as the prior distribution for $p$, we can compute the posterior distribution of $p$ by multiplying the beta prior by the binomial likelihood:

$$ P(p | y_1, y_2, ..., y_n) \propto p^{\alpha+k-1} (1-p)^{\beta+n-k-1} $$

This posterior distribution is also a beta distribution, with updated parameters $\alpha' = \alpha+k$ and $\beta'=\beta+n-k$. This means that the beta distribution is a conjugate prior to the binomial distribution, making the updating process easy to compute and interpret in this case.
